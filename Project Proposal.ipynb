# Proposal For Final Project


## 1. Project Title

Analysis of Job Title and Job Description of Labour Market in Jobstreet Malaysia


## 2. Problem Statement


A Job title should convey the information such as the domain and the responsibility of a given job in a job description. The purpose of job title is also act as a short form description that contain informations of a particular job which enable a person to communicate their roles and responsibility to other party in a clear manner.

However, this is not always the case. In popular online job advertising website in Malaysia, Jobstreet, this can be seen in a very simple search. Below are some of the examples which illustrate this case.

![jobtitle1](images/Jobtitle_1.png)
![jobdesc2](images/JobDesc1.png)

In this example, we can clearly see that the job title advertised although it describe the job roles, however if a candidate search by a keyword, in a perfect situation, the candidate would search for "Web Developer" instead of "Web Development"

![jobtitle2](images/Jobtitle_2.png)
![jobdesc2](images/JobDesc2.png)

In this example the job title goes on to include the working hours as well as the location of a given job and this would make it hard for a candidate to search for this particular job.

![jobtitle3](images/Jobtitle_3.png)
![jobdesc3](images/JobDesc3.png)

This example, the job title describe the certificate which is the Certified Environmental Professional In Scrubber Operation. It should however provide the short description of the particular role.

This project will try to answer the questions below :

1. What are the current trend of labour market across Malaysia? 
2. How are the job title corelates with the job description?
3. How best to classify and standardise the job title based on its description? 


## 3. Objective

The main objectives are :

1. To investigate the relationship between job titles and job description.
2. To build apropriate model to best classify the job titles based on its description.

## 6. Data Source

The data will be collected via webscraping from www.jobstreet.com.my which is a popular online job adverstistment among jobseekers. Web scraping is perform using combination of Selenium and BeautifulSoup in Python. Sample of data and analysis is included in the repositories. However this sample is only based on Fresh Graduate in Penang. The analysis of sample only answer few exploratory question such as :

1. What are the top ten industry that advertise their job for Fresh Graduate in Penang?
2. What are the top ten job category that are available for Fresh Graduate in Penang?
3. What is the minimum starting salary and the maximum starting salary that employers willing to pay for Fresh Graduate in Penang.


This pilot project has proved that webscraping within Jobstreet are feasible. However, cost efective method for webscraping to gather data which are suitable for modelling are in the develoment stages.
